Import library

```python
import pandas as pd
```

```python
sales=pd.read_csv("./datasets/sales.csv")
sales.head()
```

```python
items=pd.read_csv("./datasets/items.csv")
items.head()
```

```python
df=sales.copy()
df['name']=df['item_id'].map(items.set_index('id')['name'].str.lower().str.replace(" ", "_"))
df['kcal']=df['item_id'].map(items.set_index("id")["kcal"])

df.head()
```

```python
df['date']=pd.to_datetime(df['date'])
```

```python
df['month'] = df['date'].dt.month
df['year'] = df['date'].dt.year
df['day'] = df['date'].dt.day

df.head()
```

```python
df.dropna()
```

```python
df.to_csv("./datasets/cleaned_data.csv")
```
