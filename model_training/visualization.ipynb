Import Libs

```python
import pandas as pd;
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("./datasets/cleaned_data.csv")
df.head()
```

```python
monthly_sales = df.groupby(['month', 'name'])['item_count'].sum().reset_index()
daily_sales=df.groupby(['date', 'name'])['item_count'].sum().reset_index()
```

```python
daily_sales.head()
```

```python
plt.figure(figsize=(16,8))
sns.lineplot(data=monthly_sales, x="item_count", y="month", hue="name")
plt.title("Monthly Sales of Each Item")
```

```python
plt.figure(figsize=(16,8))
sns.lineplot(data=daily_sales, x="item_count", y="date", hue="name")
plt.title("Daily Sales of Each Item")
```
